```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('ticks')
sns.set_context('talk')
```

```python
# Set the seed as last four digits of matriculation number 408092
np.random.seed(8092)

# Number of particles
N_part = 10000

# Number of steps
N = 1000
```

```python
# Initialize the positions of the particles at the origin
positions = np.zeros(N_part)

# Initialize the array to store the variances
variances = np.zeros(N)

# Loop over the number of steps
for i in range(N):
    # Update the positions of the particles
    positions += np.random.choice([-1, 1], size=N_part) # step to the left or right with equal probability
    
    # Compute the variance of the positions
    variances[i] = np.mean(positions**2) - np.mean(positions)**2
```

```python
# Plot the variance of the positions

fig, ax = plt.subplots(1, 1, figsize=(8,6))

# Plot the simulated variance
ax.plot(variances, label='Simulation', zorder=1)

# Plot the theoretical variance
ax.plot(np.arange(1, N+1), label='Theory', linestyle='--', color='r', zorder=0)

ax.set_xlabel(r'Step $n$')
ax.set_ylabel(r'$\mathrm{Var}\left( x_n \right)$')

ax.set_title('Variance of the particle positions in 1D random walk')

ax.grid()
ax.legend()

sns.despine()
plt.tight_layout()
plt.show()
```
